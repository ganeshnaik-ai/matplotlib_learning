```python
data = {
    'name': [
        'Laptop', 'Mouse', 'Keyboard', 'Monitor', 'USB Hub', 
        'Headphones', 'Webcam', 'Desk Lamp', 'HDMI Cable', 'Backpack',
        'Tablet', 'Smartphone', 'Smartwatch', 'Printer', 'Router',
        'Hard Drive', 'Flash Drive', 'Cooling Pad', 'Speaker', 'Microphone'
    ],
    'price': np.random.randint(500, 50000, size=20),      # Random prices between 500 and 50,000
    'quantity': np.random.randint(1, 15, size=20),       # Random quantity between 1 and 15
    'discount': np.random.choice([5, 10, 15, 20], size=20) # Random discount percentages
}
store_df = pd.DataFrame(data)
```

```python
store_df
```

```python
store_df['totalprice']=store_df['price']*store_df['quantity']
```

```python
store_df['payable']=store_df['totalprice']-(store_df['discount']/100)*store_df['totalprice']
```

```python
store_df
```

```python
top_10=store_df.head(10)
```

```python
plt.pie(top_10['quantity'],labels=top_10['name'],autopct='%1.1f%%',startangle=140)
plt.title("Quantity Distribution of top 10 products")
plt.show()
```

```python
plt.bar(top_10['name'], top_10['payable'], color='skyblue', edgecolor='navy')
plt.title('Total Payable Amount per Product (Top 10)')
plt.xlabel('Product Name')
plt.ylabel('Payable Price')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
plt.scatter(store_df['price'], store_df['quantity'], color='purple', alpha=0.5)
plt.title('Price vs. Quantity Correlation')
plt.xlabel('Price')
plt.ylabel('Quantity')
plt.show()
```

```python
plt.hist(store_df['price'], bins=5, color='orange', edgecolor='black')
plt.title('Distribution of Product Prices')
plt.xlabel('Price Range')
plt.ylabel('Number of Products')
plt.show()
```

```python
plt.boxplot(store_df['payable'])
plt.title('Payable Amount Summary (Outlier Detection)')
plt.ylabel('Price')
plt.show()
```

Data Type	Goal	Recommended Plot
Trend over Time	See how values change	Line Plot (plt.plot)
Categorical	Compare values between groups	Bar Chart (plt.bar)
Proportions	See parts of a whole (percentage)	Pie Chart (plt.pie)
Relationships	See if X affects Y	Scatter Plot (plt.scatter)
Distribution	See the frequency of values	Histogram (plt.hist)
Statistics	Find the median and outliers	Box Plot (plt.boxplot)
